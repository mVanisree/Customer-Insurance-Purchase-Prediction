import libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

import Dataset 

```python
df=pd.read_csv('Social_Network_Ads.csv')
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

Splitting the Dataset into training and testing set 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

Feature Scaling 

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
#Always use fit_transform() on training data, and
#Only use transform() on test/validation data.
```

```python
print(x_train)
```

```python
print(x_test)
```

Training the logistic regression model on the training set 

```python
from sklearn.linear_model import LogisticRegression 
classifier=LogisticRegression()
classifier.fit(x_train,y_train)
```

predicting a new result 


```python
print(classifier.predict(sc.transform([[30,87000]])))
```

```python
print(classifier.predict(sc.transform([[45,26000]])))
```

```python
print(classifier.predict(sc.transform([[62,120000]])))
```

predicting test set results 

```python
y_pred = classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

Confusion matrix

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

Visualizing the trainign set results

```python
from matplotlib.colors import ListedColormap
x_set,y_set=sc.inverse_transform(x_train),y_train
x1,x2=np.meshgrid(np.arange(start=x_set[:,0].min()-10,stop=x_set[:,0].max()+10,step=0.55)
                  ,np.arange(start=x_set[:,1].min()-1000,stop=x_set[:,1].max()+1000,step=0.55))
plt.contourf(x1, x2,classifier.predict(sc.transform(np.array([x1.ravel(), x2.ravel()]).T)).reshape(x1.shape),alpha=0.75,cmap=ListedColormap(('red', 'green')))
plt.xlim(x1.min(),x1.max())
plt.ylim(x2.min(),x2.max())
for i ,j in enumerate (np.unique(y_set)):
    plt.scatter(x_set[y_set==j,0],x_set[y_set==j,1],c=ListedColormap(('red','green'))(i),label=j)
plt.title("logistic regression('training set')")
plt.xlabel('age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()       
```

```python
from matplotlib.colors import ListedColormap
x_set,y_set=sc.inverse_transform(x_test),y_test
x1,x2=np.meshgrid(np.arange(start=x_set[:,0].min()-10,stop=x_set[:,0].max()+10,step=0.55)
                  ,np.arange(start=x_set[:,1].min()-1000,stop=x_set[:,1].max()+1000,step=0.55))
plt.contourf(x1, x2,classifier.predict(sc.transform(np.array([x1.ravel(), x2.ravel()]).T)).reshape(x1.shape),alpha=0.75,cmap=ListedColormap(('red', 'green')))
plt.xlim(x1.min(),x1.max())
plt.ylim(x2.min(),x2.max())
for i ,j in enumerate (np.unique(y_set)):
    plt.scatter(x_set[y_set==j,0],x_set[y_set==j,1],c=ListedColormap(('red','green'))(i),label=j)
plt.title("logistic regression('testing set')")
plt.xlabel('age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show() 
```
